```python
import pandas as pd
from pathlib import Path
import IMP 
import IMP.rmf
import RMF
```

```python
analysis_dir = Path(Path.home(), "mtorc2/exp_9/71/analysis/analysis.test")
rmf_file = Path(analysis_dir, "sampcon/cluster.0/cluster_center_model.rmf3")
fh = RMF.open_rmf_file_read_only(str(rmf_file))
m = IMP.Model()
h = IMP.rmf.create_hierarchies(fh, m)[0]
IMP.rmf.load_frame(fh, 0)
```

```python
pdb_file = Path(analysis_dir, "sampcon/cluster.0/cluster_center_model.pdb")
fixed_file = Path(analysis_dir, "cluster_center_model.pdb")
IMP.atom.write_pdb(h, str(pdb_file))
```

```python
max_res_id_dict = {"A": 2549, "B": 1695, "C": 324, "D": 155, "E": 121, "F": 99, "G": 140, "H": 141, "I": 333, "J": 2549, "K": 1695, "L": 324, "M": 155, "N": 121, "O": 99, "P": 140, "Q": 141, "R": 333}
            
```

```python
f_in = open(pdb_file, "r")
f_out = open(fixed_file, "w")
f_in.readline()
x = f_in.readline()
chain = "A" 
while "ENDMDL" not in x: 
        
    if "UNK" in x: 
        x = x[:21] + chain + x[22:]
        x = x[:17] + "BEA" + x[20:]        

    print(x)
    f_out.write(x)
    
    res_id = int(x[22:26])     
    if res_id == max_res_id_dict[chain]:
        if ord(chain) < 74: 
            chain = chr(ord(chain) + 9)
        else: 
            chain = chr(ord(chain) - 8)
        
    x = f_in.readline()

f_out.write("ENDMDL")
f_in.close()
f_out.close()
```
