```python
# ref_pdb_file = Path(Path.home(), "mtorc2/manuscript/submission_2/models/126_0_1/cluster.0/cluster_center_model.pdb")

ref_pdb_file = Path(Path.home(), "mtorc2/manuscript/submission_2/models/126_0_1/cluster.0/cluster_center_model.rmf3")
fh_ref = RMF.open_rmf_file_read_only(str(ref_pdb_file))
m_ref = IMP.Model()
h_ref = IMP.rmf.create_hierarchies(fh_ref, m_ref)[0]
```

```python
sel_ref = IMP.atom.Selection(h_ref, resolution=10, chain_id="E", residue_index=292)
pids_ref = [pid for pid in sel_ref.get_selected_particle_indexes()]
xyzs_ref = [IMP.core.XYZ(m_ref, pid) for pid in pids_ref]
print(len(xyzs_ref))
```

```python
del fh_edit
```

```python
gsms_edit_dir = Path("/wynton/group/sali/mhancock/mtorc2/samples/exp_14/129/analysis/1/sampcon_3/gsms/10000R_edit/")
shutil.rmtree(gsms_edit_dir)
gsms_edit_dir.mkdir()

for file in ["A", "B"]: 
    print(file)
    rmf_file_edit = Path(gsms_edit_dir, "{}.rmf3".format(file))
    fh_edit = RMF.create_rmf_file(str(rmf_file_edit))

    rmf_file = Path(gsms_dir, "{}.rmf3".format(file))
    fh = RMF.open_rmf_file_read_only(str(rmf_file))
    
    txt_file_edit = Path(gsms_edit_dir, "{}.txt".format(file))
    fh_txt_out = open(txt_file_edit, "w")
    
    txt_file = Path(gsms_dir, "{}.txt".format(file))
    fh_txt = open(txt_file, "r")
    
    m = IMP.Model()
    h = IMP.rmf.create_hierarchies(fh, m)[0]

    IMP.rmf.add_hierarchy(fh_edit, h)

    for i in range(10000):
        IMP.rmf.load_frame(fh, RMF.FrameID(i))
        score = float(fh_txt.readline())
        
        sel_0 = IMP.atom.Selection(h, resolution=10, residue_index=292, chain_id="E")
        pids_0 = [pid for pid in sel_0.get_selected_particle_indexes()]
        xyzs_0 = [IMP.core.XYZ(m, pid) for pid in pids_0]

        sel_1 = IMP.atom.Selection(h, resolution=10, residue_index=292, chain_id="J")
        pids_1 = [pid for pid in sel_1.get_selected_particle_indexes()]
        xyzs_1 = [IMP.core.XYZ(m, pid) for pid in pids_1]
        
        rmsd_0 = IMP.atom.get_rmsd(xyzs_ref, xyzs_0)
        rmsd_1 = IMP.atom.get_rmsd(xyzs_ref, xyzs_1)
        
        if rmsd_1 > rmsd_0: 
            print(i, score, rmsd_0, rmsd_1)
            IMP.rmf.save_frame(fh_edit)
            fh_txt_out.write(str(score) + "\n")
                    
    fh_txt_out.close()

    del fh
    del fh_edit
            
```
