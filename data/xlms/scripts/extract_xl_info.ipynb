```python
import pandas as pd
```

```python
xl_file = Path(Path.home(), "Documents/mtorc2/data/xlms/unformatted/CX_summary_new")
xl_df = pd.read_csv(xl_file)
```

```python
xl_df = xl_df.dropna(subset=["Proteins"])
na_columns = list(range(12,25))
xl_df = xl_df.drop(xl_df.columns[na_columns], axis=1)
print(len(xl_df))
xl_df.head()
```

```python
for i in range(len(xl_df)):
    if i < 37: 
        xl_df.iloc[i, xl_df.columns.get_loc("Linker")] = "DSS"
    else: 
        xl_df.iloc[i, xl_df.columns.get_loc("Linker")] = "EDC"
        
print(len(xl_df))
xl_df.head()
```

```python
res1s, res2s, prot1s, prot2s = list(), list(), list(), list() 
for i in range(len(xl_df)):
    entry = xl_df.iloc[i, xl_df.columns.get_loc("Proteins")]
    
    prot1 = entry.split("-")[0].split("(")[0].upper()
    prot2 = entry.split("-")[1].split("(")[0].upper()
    
    if prot1 == "RICTR": 
        prot1 = "RICTOR"
    if prot2 == "RICTR": 
        prot2 = "RICTOR"
    
    res1 = entry.split("-")[0].split("(")[1].split(")")[0]
    res2 = entry.split("-")[1].split("(")[1].split(")")[0]
    
    res1s.append(int(res1))
    res2s.append(int(res2))
    prot1s.append(prot1)
    prot2s.append(prot2)
    
xl_df["res1"] = res1s
xl_df["res2"] = res2s
xl_df["prot1"] = prot1s
xl_df["prot2"] = prot2s
print(len(xl_df))
xl_df.head()
```

```python
rb_xls = list()
rb_xls.append(("MLST8", 86, "MTOR", 2374))
rb_xls.append(("MTOR", 2113, "RICTOR", 274))
rb_xls.append(("RICTOR", 699, "MTOR", 84))
rb_xls.append(("MTOR", 1766, "RICTOR", 586))
rb_xls.append(("MTOR", 2113, "RICTOR", 270))
rb_xls.append(("RICTOR", 126, "MTOR", 2370))
```

```python
rbs = list()
for i in range(len(xl_df)): 
    prot1 = xl_df.iloc[i, xl_df.columns.get_loc("prot1")]
    prot2 = xl_df.iloc[i, xl_df.columns.get_loc("prot2")]
    res1 = xl_df.iloc[i, xl_df.columns.get_loc("res1")]
    res2 = xl_df.iloc[i, xl_df.columns.get_loc("res2")]
    
    rb_xl = False
    for j in range(len(rb_xls)):
        if prot1 == rb_xls[j][0] and res1 == rb_xls[j][1] and prot2 == rb_xls[j][2] and res2 == rb_xls[j][3]:
            rb_xl = True 
        elif prot1 == rb_xls[j][2] and res1 == rb_xls[j][3] and prot2 == rb_xls[j][0] and res2 == rb_xls[j][1]:
            rb_xl = True 
            
        if rb_xl: 
            break
    
    rbs.append(rb_xl)
    
xl_df["rbs"] = rbs
print(len(xl_df))
xl_df.head()
```

### Split dataframes into xl_rb_dss, xl_rb_edc, xl_flex_dss, xl_flex_edc

```python
rb_dss = list() 
rb_edc = list()
flex_dss = list() 
flex_edc = list()

for i in range(len(xl_df)):   
    prot1 = xl_df.iloc[i, xl_df.columns.get_loc("prot1")]
    prot2 = xl_df.iloc[i, xl_df.columns.get_loc("prot2")]
    res1 = xl_df.iloc[i, xl_df.columns.get_loc("res1")]
    res2 = xl_df.iloc[i, xl_df.columns.get_loc("res2")]
    
    linker = xl_df.iloc[i, xl_df.columns.get_loc("Linker")]
    rb = xl_df.iloc[i, xl_df.columns.get_loc("rbs")]
    if linker == "DSS" and rb:
        rb_dss.append([prot1, res1, prot2, res2])
    elif linker == "DSS" and not rb: 
        flex_dss.append([prot1, res1, prot2, res2])
    elif linker == "EDC" and rb: 
        rb_edc.append([prot1, res1, prot2, res2])
    elif linker == "EDC" and not rb: 
        flex_edc.append([prot1, res1, prot2, res2])
        
```

```python
rb_dss_df = pd.DataFrame(data=rb_dss, columns=["prot1", "res1", "prot2", "res2"])
rb_dss_df.to_csv("/Users/matthew/Documents/mtorc2/data/xlms/inter_rb_dss.csv", sep=",")
print(len(rb_dss_df))
rb_dss_df.head()
```

```python
flex_dss_df = pd.DataFrame(data=flex_dss, columns=["prot1", "res1", "prot2", "res2"])
flex_dss_df.to_csv("/Users/matthew/Documents/mtorc2/data/xlms/inter_flex_dss.csv", sep=",")
print(len(flex_dss_df))
flex_dss_df.head()
```

```python
rb_edc_df = pd.DataFrame(data=rb_edc, columns=["prot1", "res1", "prot2", "res2"])
rb_edc_df.to_csv("/Users/matthew/Documents/mtorc2/data/xlms/inter_rb_edc.csv", sep=",")
print(len(rb_edc_df))
rb_edc_df.head()
```
