```python
import pandas as pd
```

```python
xl_df = pd.read_csv("/Users/matthew/Documents/mtorc2/data/xlms/all_xl.csv")
xl_df = xl_df.drop(columns=["Unnamed: 0"])

```

```python
for i in range(len(xl_df)):
    print(tuple(xl_df.iloc[i]))
```

```python
for i in range(len(xl_df)):
    prot1, res1, prot2, res2 = tuple(xl_df.iloc[i])

    if prot1 == "MSIN1" and (res1 >= 139 and res1 < 279): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "CRIM"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1 - 132 + 1 
    if prot2 == "MSIN1" and (res1 >= 139 and res1 < 279): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "CRIM"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2 - 132 + 1        
    if prot1 == "MSIN1" and (res1 >= 279 and res1 < 377): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "RBD"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1 - 279 + 1
    if prot1 == "MSIN1" and (res2 >= 279 and res2 < 377): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "RBD"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2 - 279 + 1
    if prot1 == "MSIN1" and res1 >= 377: 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "PH"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1 - 377 + 1  
    if prot2 == "MSIN1" and res2 >= 377: 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "PH"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2 - 377 + 1             
            
            
            
```
