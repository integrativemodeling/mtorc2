```python
import pandas as pd
from pathlib import Path
```

```python
xl_sat_file = Path(Path.home(), "Documents/mtorc2/data/xlms/all_xl.csv")
xl_df = pd.read_csv(xl_sat_file)
xl_df = xl_df.drop(columns=["Unnamed: 0"])
xl_df.head() 
```

```python
target_xls = list() 
for i in range(len(xl_df)):
    prot1, res1, prot2, res2 = tuple(xl_df.iloc[i])
        
    if (prot1 == "RICTOR" or prot1 == "MTOR") and prot2 == "AKT1": 
        target_xls.append(i) 
    if (prot2 == "RICTOR" or prot2 == "MTOR") and prot1 == "AKT1": 
        target_xls.append(i)

```

```python
print(len(target_xls))
```

```python
xl_df = xl_df.iloc[target_xls].copy() 
# df_copy = df_copy.drop(columns=["contains", "satisfied"])
for i in range(len(xl_df)): 
    print(tuple(xl_df.iloc[i]))
```

```python
xl_df.to_csv(Path(Path.home(), "Documents/mtorc2/data/xlms/akt.rictor.mtor.csv"))
```
