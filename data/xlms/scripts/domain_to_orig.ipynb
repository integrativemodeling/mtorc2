```python
import pandas as pd
from pathlib import Path
```

```python
xl_file = Path(Path.home(), "Documents/mtorc2/manuscript/submission/models/89_2462_1_8525/xls.csv")
xl_df = pd.read_csv(xl_file)
xl_df = xl_df.drop(columns=["Unnamed: 0"])
xl_df.head()
```

```python
def correct_entry(prot, res):
    corr_dir = dict()
    corr_dir["CRIM"] = ("MSIN1", 157)
    corr_dir["RBD"] = ("MSIN1", 279)
    corr_dir["MSIN1PH"] = ("MSIN1", 381)
    corr_dir["AKT1PH"] = ("AKT1", 0)
    corr_dir["KINASE"] = ("AKT1", 145) 
    
    prot_corr, res_corr = prot, res 
    if prot in list(corr_dir.keys()): 
        prot_corr, offset = corr_dir[prot]
        res_corr = res + offset 
        
    return prot_corr, res_corr 
```

```python
for i in range(len(xl_df)):
    prot_1, res_1, prot_2, res_2 = xl_df.iloc[i, 0:4]
    prot_corr_1, res_corr_1 = correct_entry(prot_1, res_1)
    prot_corr_2, res_corr_2 = correct_entry(prot_2, res_2)

    print(prot_corr_1, res_corr_1, prot_corr_2, res_corr_2)
    xl_df.iloc[i, 0:4] = prot_corr_1, res_corr_1, prot_corr_2, res_corr_2
    
```

```python
xl_corr_file = Path(Path.home(), "Documents/mtorc2/manuscript/submission/models/89_2462_1_8525/xls_corr.csv")
xl_df.to_csv(xl_corr_file)

```
