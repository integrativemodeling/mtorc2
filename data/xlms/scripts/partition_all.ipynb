```python
from pathlib import Path
import pandas as pd
```

```python
xl_file = Path(Path.home(), "Documents/mtorc2/data/xlms/csvs/all.csv")
xl_df = pd.read_csv(xl_file)
xl_df.head()
```

```python
xls_dict = dict() 
xls_dict["dss"] = list() 
xls_dict["edc"] = list() 

for i in range(len(xl_df)): 
    res1, prot1, res2, prot2, xl_type = xl_df.iloc[i, 0:5]
    print(res1, prot1, res2, prot2, xl_type)
    
    if xl_type == "dss": 
        xls_dict["dss"].append(i)
    elif xl_type == "edc": 
        xls_dict["edc"].append(i)
    
    
```

```python
len(xls_dict["dss"]) + len(xls_dict["edc"]) - len(xl_df)
```

```python
for key in xls_dict.keys(): 
    part_df = xl_df.iloc[xls_dict[key]]
    print(len(part_df))
    part_file = Path(Path.home(), "Documents/mtorc2/data/xlms/csvs/{}.csv".format(key))
    part_df.to_csv(part_file)
```
