```python
import pandas as pd
```

```python
xl_file = "/Users/matthew/Documents/mtorc2/data/xlms/xl_intra.csv"
xl_df = pd.read_csv(xl_file)
```

```python
xl_df = xl_df.dropna(subset=["Proteins"])
```

```python
xl_df.head()
```

```python
entries = list() 
for entry in xl_df["Proteins"]: 
    prot1 = entry.split("-")[0].split("(")[0].upper()
    prot2 = entry.split("-")[1].split("(")[0].upper()
    
    if prot1 == "RICTR": 
        prot1 = "RICTOR"
    if prot2 == "RICTR": 
        prot2 = "RICTOR"
    
    res1 = entry.split("-")[0].split("(")[1].split(")")[0]
    res2 = entry.split("-")[1].split("(")[1].split(")")[0]
    
    entries.append([prot1, res1, prot2, res2])
```

```python
xl_df_new = pd.DataFrame(data=entries, columns=["prot1", "res1", "prot2", "res2"])
xl_df_new.head()
```

```python
xl_df_new.to_csv("/Users/matthew/Documents/mtorc2/data/xlms/xl_intra_dss_formatted.csv", sep=",")
```
