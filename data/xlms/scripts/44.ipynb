```python
import pandas as pd
from pathlib import Path
```

```python
xl_sat_file = Path(Path.home(), "Documents/mtorc2/data/xlms/all_xl.csv")
xl_df = pd.read_csv(xl_sat_file)
xl_df = xl_df.drop(columns=["Unnamed: 0"])
xl_df.head() 
```

```python
# Flip MTOR so that it is always prot1. 
for i in range(len(xl_df)): 
    if xl_df.iloc[i, xl_df.columns.get_loc("prot2")] == "MTOR": 
        prot1_copy = xl_df.iloc[i, xl_df.columns.get_loc("prot1")]
        res1_copy = xl_df.iloc[i, xl_df.columns.get_loc("res1")]
        
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = xl_df.iloc[i, xl_df.columns.get_loc("prot2")]
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = xl_df.iloc[i, xl_df.columns.get_loc("res2")]

        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = prot1_copy
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res1_copy
```

```python
xl_df.head()
```

```python
target_xls = list() 
for i in range(len(xl_df)):
    prot1, res1, prot2, res2 = tuple(xl_df.iloc[i])

    if prot1 == "MSIN1" and (res1 >= 158 and res1 < 280): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "CRIM"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1-158+1 
    if prot2 == "MSIN1" and (res2 >= 158 and res2 < 280): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "CRIM"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2-158+1   
        
    if prot1 == "MSIN1" and (res1 >= 280 and res1 < 382): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "RBD"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1-280+1
    if prot2 == "MSIN1" and (res2 >= 280 and res2 < 382): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "RBD"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2-280+1
        
    if prot1 == "MSIN1" and (res1 >= 382): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "MSIN1PH"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1-382+1
    if prot2 == "MSIN1" and (res2 >= 382): 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "MSIN1PH"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2-382+1
        
    if prot1 == "AKT1" and res1 < 146: 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "AKT1PH"
    if prot2 == "AKT1" and res2 < 146: 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "AKT1PH"
        
    if prot1 == "AKT1" and res1 >= 146: 
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = "KINASE"
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = res1-146+1 
    if prot2 == "AKT1" and res2 >= 146: 
        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = "KINASE"
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res2-146+1              

#     if prot1 in ["RICTOR", "MTOR"] and prot2 in ["AKT1"]: 
#         continue 
#     if prot2 in ["RICTOR", "MTOR"] and prot1 in ["AKT1"]: 
#         continue
#     if prot1 in ["AKT1"] and prot2 in ["AKT1"]: 
#         continue  

    target_xls.append(i)

```
