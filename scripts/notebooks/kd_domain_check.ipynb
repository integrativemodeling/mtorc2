### Iterate through PH domains in the 2 

```python
import pandas as pd
from pathlib import Path
import IMP 
import IMP.atom
import IMP.rmf
import RMF
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
gsms_dir = Path(Path.home(), "mtorc2/exp_10/89/analysis/2462/sampcon_1/gsms/5000")
gsms_file = Path(gsms_dir, "A.rmf3")
fh_clust = RMF.open_rmf_file_read_only(str(gsms_file))
m_clust = IMP.Model()
h_clust = IMP.rmf.create_hierarchies(fh_clust, m_clust)[0]
    
```

```python
s = IMP.atom.Selection(h_clust, chain_id="R", resolution=10)
print(len(s.get_selected_particles()))
```

```python
df_data = {"frame": list(), "d_A": list(), "d_A_alt": list(), "d_active": list(), "d_active_alt": list()}
for i in range(5000):
    if i % 100 == 0: 
        print(i)
    
    IMP.rmf.load_frame(fh_clust, RMF.FrameID(i))

    s_1 = IMP.atom.Selection(h_clust, resolution=10, residue_indexes=list(range(90,272)), chain_id="I")
    s_2 = IMP.atom.Selection(h_clust, resolution=10, residue_indexes=list(range(90,272)), chain_id="R")

    pids_1 = [pid for pid in s_1.get_selected_particle_indexes()]
    pids_2 = [pid for pid in s_2.get_selected_particle_indexes()]

    xyzs_1 = [IMP.core.XYZ(m_clust, pid) for pid in pids_1]
    xyzs_2 = [IMP.core.XYZ(m_clust, pid) for pid in pids_2]

    avg_1, avg_2 = 0, 0 
    for xyz in xyzs_1: 
        d = (xyz.get_coordinates() - IMP.algebra.Vector3D(111.854, 112.456, 124.052)).get_magnitude()
        avg_1 = avg_1 + d
        
    for xyz in xyzs_2: 
        d = (xyz.get_coordinates() - IMP.algebra.Vector3D(111.854, 112.456, 124.052)).get_magnitude()
        avg_2 = avg_2 + d
        
    avg_1 = avg_1 / len(xyzs_1)
    avg_2 = avg_2 / len(xyzs_2)
    
    df_data["frame"].append(i)
    if avg_2 < avg_1: 
        df_data["d_A"].append(avg_2)
        df_data["d_A_alt"].append(1)
    else: 
        df_data["d_A"].append(avg_1)
        df_data["d_A_alt"].append(0)

    avg_1, avg_2 = 0, 0 
    for xyz in xyzs_1: 
        d = (xyz.get_coordinates() - IMP.algebra.Vector3D(122.968, 44.546, 180.969)).get_magnitude()
        avg_1 = avg_1 + d
        
    for xyz in xyzs_2: 
        d = (xyz.get_coordinates() - IMP.algebra.Vector3D(122.968, 44.546, 180.969)).get_magnitude()
        avg_2 = avg_2 + d

    avg_1 = avg_1 / len(xyzs_1)
    avg_2 = avg_2 / len(xyzs_2)
    
    if avg_2 < avg_1: 
        df_data["d_active"].append(avg_2)
        df_data["d_active_alt"].append(1)
    else: 
        df_data["d_active"].append(avg_1)
        df_data["d_active_alt"].append(0)

            
#     print(avg_1, avg_2)
    
# del fh_clust
```

```python
dist_df = pd.DataFrame(df_data)
dist_df.head()
```

```python
dist_df.nsmallest(10, "d_A")
```

```python
dist_df.nsmallest(10, "d_active")
```

```python
pdb_dir = Path(Path.home(), "mtorc2/manuscript/submission/figures/akt_atp/akt_A")
for frame_id in dist_df.nsmallest(25, "d_A")["frame"]:
    print(frame_id)
    d_A_alt = dist_df.iloc[frame_id, dist_df.columns.get_loc("d_A_alt")]
#     d_active_alt = dist_df.iloc[frame_id, dist_df.columns.get_loc("d_active_alt")]

    IMP.rmf.load_frame(fh_clust, RMF.FrameID(frame_id))

    if d_A_alt: 
        chain_id = "R"
    else: 
        chain_id = "I"
        
    sel = IMP.atom.Selection(h_clust, resolution=1, residue_indexes=list(range(90,272)), chain_id=chain_id)

    IMP.atom.write_pdb(sel, str(Path(pdb_dir, str(frame_id)+".pdb")))
    
    
```
