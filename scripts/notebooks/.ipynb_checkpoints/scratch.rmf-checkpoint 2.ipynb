```python
flex_bead_names = list() 
for p in ps_10: 
    if str(p).split('"')[1].split(" ")[0] != "Fragment": 
        print(str(p).split('"')[1])
        flex_bead_names.append(str(p).split('"')[1])
```

```python
flex_coords_dict = dict() 
mol = "" 
for pid in m.get_particle_indexes(): 
#     print(str(m.get_particle_name(pid)))
    if m.get_particle_name(pid) == "MTOR": 
        mol = "MTOR"
    if m.get_particle_name(pid) == "RICTOR": 
        mol = "RICTOR"
    if m.get_particle_name(pid) == "MLST8": 
        mol = "MLST8"
        
    if str(m.get_particle_name(pid)) in flex_bead_names: 
        d = IMP.core.XYZ(m, pid)
        print(m.get_particle_name(pid), d.get_coordinates(), mol)
        
        mol_pid = mol+"_"+m.get_particle_name(pid)
        if mol_pid not in flex_coords_dict.keys():
            flex_coords_dict[mol_pid] = d.get_coordinates()
        
```
