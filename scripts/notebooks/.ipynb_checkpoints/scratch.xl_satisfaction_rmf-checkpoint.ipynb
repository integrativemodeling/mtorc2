```python
def get_bead_rmf(m, chain_ids, prot, res, copy): 
    chain_ids_inv = dict()
    for key in chain_ids.keys(): 
        chain_ids_inv[chain_ids[key]] = key
    
    cur_prot, cur_copy = "", 0
    name, xyz = "", (0,0,0)
    
    for pid in m.get_particle_indexes(): 
        if IMP.atom.Chain.get_is_setup(m, pid): 
            chain_id = IMP.atom.Chain(m, pid).get_id()
            cur_prot, cur_copy = chain_ids_inv[chain_id].split("_")
            cur_copy = int(cur_copy)
#             print(cur_prot, cur_copy)
        
        # Check if the protein copy matches. 
#         print(m.get_particle_name(pid))

        if cur_prot == prot and cur_copy == copy and IMP.core.XYZR.get_is_setup(m, pid):  
            # Check if the residue is at resolution 10. 
            if IMP.atom.Fragment.get_is_setup(m, pid): 
                if IMP.atom.Fragment(m, pid).get_contains_residue(res): 
                    d = IMP.core.XYZR(m, pid)
                    name, xyz = m.get_particle_name(pid), d.get_coordinates()  
            # Check if the residue is at resolution 1. 
            if IMP.atom.Residue.get_is_setup(m, pid): 
                if IMP.atom.Residue(m, pid).get_index() == res:
                    d = IMP.core.XYZR(m, pid)
                    name, xyz = m.get_particle_name(pid), d.get_coordinates()  
                
    return name, xyz
    
```

```python
xl_file = Path(Path.home(), "mtorc2/data/xlms/39.csv")
xl_df = pd.read_csv(xl_file)
xl_df = xl_df.drop(columns=["Unnamed: 0"])
xl_df.head()
```

```python
# Flip MTOR so that it is always prot1. 
for i in range(len(xl_df)): 
    if xl_df.iloc[i, xl_df.columns.get_loc("prot2")] == "MTOR": 
        prot1_copy = xl_df.iloc[i, xl_df.columns.get_loc("prot1")]
        res1_copy = xl_df.iloc[i, xl_df.columns.get_loc("res1")]
        
        xl_df.iloc[i, xl_df.columns.get_loc("prot1")] = xl_df.iloc[i, xl_df.columns.get_loc("prot2")]
        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = xl_df.iloc[i, xl_df.columns.get_loc("res2")]

        xl_df.iloc[i, xl_df.columns.get_loc("prot2")] = prot1_copy
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = res1_copy
```

```python
def get_pdb_bead(bead):
    num_range = bead.split("_")[0]
    
    min_num, max_num = num_range.split("-")
    min_num = int(min_num)
    max_num = int(max_num)
    
    return round((min_num + max_num)/2 + .1)
        
```

```python
contains = list() 
copies1 = list() 
copies2 = list() 
xl_sats = list() 
for i in range(len(xl_df)): 
    prot1, res1, prot2, res2 = xl_df.iloc[i,0:4]
    res1 = int(res1)
    res2 = int(res2)
#     bead1_0, coord1_0 = get_bead(m, chain_ids, prot1, res1, 0)
#     bead1_1, coord1_1 = get_bead(m, chain_ids, prot1, res1, 1)
#     bead2_0, coord2_0 = get_bead(m, chain_ids, prot2, res2, 0)
#     bead2_1, coord2_1 = get_bead(m, chain_ids, prot2, res2, 1)
    
#     if len(bead1_0.split("_")) > 1: 
#         print(bead1_0, get_pdb_bead(bead1_0))
#         xl_df.iloc[i, xl_df.columns.get_loc("res1")] = get_pdb_bead(bead1_0)
#     if len(bead2_0.split("_")) > 1:
#         print(bead2_0, get_pdb_bead(bead2_0))
#         xl_df.iloc[i, xl_df.columns.get_loc("res2")] = get_pdb_bead(bead2_0)

    coord1_0 = get_coords(atoms_dict, prot1, res1, 0)
    coord1_1 = get_coords(atoms_dict, prot1, res1, 1)
    coord2_0 = get_coords(atoms_dict, prot2, res2, 0)
    coord2_1 = get_coords(atoms_dict, prot2, res2, 1)

    contain = False  
    xl_sat = False 
    l2norm_0, l2norm_1 = 0,0 
    copy1, copy2 = 0,0
    if bead1_0 and bead2_0: 
        contain = True   

        xl_df.iloc[i, xl_df.columns.get_loc("res1")] = int(bead1_0)
        xl_df.iloc[i, xl_df.columns.get_loc("res2")] = int(bead2_0)    

        l2norm_0 = np.linalg.norm(np.array(coord1_0)-np.array(coord2_0))
        l2norm_1 = np.linalg.norm(np.array(coord1_0)-np.array(coord2_1))
        
        if (l2norm_0 < 35) or (l2norm_1 < 35): 
            xl_sat = True 
        
    if l2norm_1 < l2norm_0: 
        copy2 = 1 
        
    copies1.append(copy1)
    copies2.append(copy2)
    xl_sats.append(xl_sat)
    contains.append(contain)
    
    print(i, prot1, res1, prot2, res2, contain, xl_sat, l2norm_0, l2norm_1, copy1, copy2)
    print(coord1_0, coord2_0, coord2_1)
    
xl_df["contains"] = contains
xl_df["xl_sat"] = xl_sats
xl_df["copy_1"] = copies1
xl_df["copy_2"] = copies2
```

```python
sat_xls = str()
unsat_xls = str()
for i in range(len(xl_df)):
    if xl_df.iloc[i, xl_df.columns.get_loc("contains")]:
        res1 = xl_df.iloc[i, xl_df.columns.get_loc("res1")]
        res2 = xl_df.iloc[i, xl_df.columns.get_loc("res2")]

        copy1 = xl_df.iloc[i, xl_df.columns.get_loc("copy_1")]
        copy2 = xl_df.iloc[i, xl_df.columns.get_loc("copy_2")]

        chain1 = chain_ids[xl_df.iloc[i, xl_df.columns.get_loc("prot1")]+"_"+str(copy1)]
        chain2 = chain_ids[xl_df.iloc[i, xl_df.columns.get_loc("prot2")]+"_"+str(copy2)]
        
        if res1 == res2 and prot1 == prot2: 
#         print("SELF", prot1, res1, prot2, res2)
            continue
    
        if xl_df.iloc[i, xl_df.columns.get_loc("xl_sat")]:
            sat_xls = sat_xls + "#1/{}:{} #1/{}:{}\n".format(chain1, res1, chain2, res2)
        else: 
            unsat_xls = unsat_xls + "#1/{}:{} #1/{}:{}\n".format(chain1, res1, chain2, res2)

print(sat_xls)
print
print(unsat_xls)
```

```python
n_contains = dict()
n_satisfied = dict() 
n_xls = dict() 

prots = ["MTOR", "RICTOR", "MSIN1", "MLST8", "AKT1"]
for prot in prots: 
    for prot2 in prots: 
        entry = prot + "|" + prot2
        if prot2 + "|" + prot not in n_contains.keys(): 
            n_contains[entry] = 0
            n_satisfied[entry] = 0
            n_xls[entry] = 0

for i in range(len(xl_df)): 
    prot1 = xl_df.iloc[i, xl_df.columns.get_loc('prot1')]
    prot2 = xl_df.iloc[i, xl_df.columns.get_loc('prot2')]
    
    if prot1 == "CRIM" or prot1 == "RBD": 
        prot1 = "MSIN1"
    if prot2 == "CRIM" or prot2 == "RBD": 
        prot2 = "MSIN1"
    
    entry = prot1 + "|" + prot2
    if entry not in n_contains.keys(): 
        entry = prot2 + "|" + prot1

    n_xls[entry] = n_xls[entry] + 1

    if xl_df.iloc[i, xl_df.columns.get_loc('contains')]: 
        n_contains[entry] = n_contains[entry] + 1
        
    if xl_df.iloc[i, xl_df.columns.get_loc('xl_sat')]:
        n_satisfied[entry] = n_satisfied[entry] + 1 
        
fig, ax = plt.subplots(figsize=(25, 18), dpi=80) 
labels = list(n_contains.keys())
contains_list = [n_contains[key] for key in n_contains.keys()]
satisfied_list = [n_satisfied[key] for key in n_contains.keys()]
n_xls_list = [n_xls[key] for key in n_contains.keys()]

x = np.arange(len(labels))  # the label locations
width = .75  # the width of the bars

rects3 = ax.bar(x, n_xls_list, width/3, label='Total')
rects1 = ax.bar(x + width/3, contains_list, width/3, label='Contains')
rects2 = ax.bar(x + 2*width/3, satisfied_list, width/3, label='Satisfied')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Counts')
ax.set_title('XL Counts')
ax.set_xticks(x, labels)
ax.set_xticklabels(labels, rotation=45)
ax.legend()

ax.bar_label(rects1, padding=1)
ax.bar_label(rects2, padding=1)
ax.bar_label(rects3, padding=1)

fig.tight_layout()
# plt.show()
plt.savefig("/home/matthew/Downloads/plot.png")
```
