```python
import pandas as pd
from pathlib import Path
import IMP 
import IMP.rmf
import RMF
```

```python
rmf_file = Path(Path.home(), "mtorc2/exp_8/66/analysis/analysis.413611/sampcon30/cluster.0/cluster_center_model.rmf3")
fh = RMF.open_rmf_file_read_only(str(rmf_file))
m = IMP.Model()
h = IMP.rmf.create_hierarchies(fh, m)[0]
IMP.rmf.load_frame(fh, 0)
```

```python
pids = IMP.atom.Selection(h, molecule="MSIN1", resolution=10).get_selected_particle_indexes()
```

```python
pdb_file = Path(Path.home(), "mtorc2/exp_8/66/analysis/analysis.413611/sampcon30/cluster.0/cluster_center_model.pdb")
IMP.atom.write_pdb(h, str(pdb_file))
```

```python
fixed_file = Path(Path.home(), "mtorc2/exp_8/66/analysis/analysis.413611/sampcon30/cluster.0/cluster_center_model.fixed.pdb")
```

```python
max_res_id_dict = {"A": 2549, "B": 1695, "C": 324, "D": 155, "E": 121, "F": 99, "G": 140, "H": 141, "I": 333}
```

```python
f_in = open(pdb_file, "r")
f_out = open(fixed_file, "w")
f_in.readline()
x = f_in.readline()
chain = "A" 
while "ENDMDL" not in x: 
        
    if "UNK" in x: 
        x = x[:21] + chain + x[22:]
        x = x[:17] + "BEA" + x[20:]        

    print(x)
    f_out.write(x)
    
    res_id = int(x[22:26])     
    if res_id == max_res_id_dict[chain]: 
        chain = chr(ord(chain) + 1)
        
    x = f_in.readline()

f_out.write("ENDMDL")
f_in.close()
f_out.close()
```
