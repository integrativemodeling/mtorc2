### Check to see if there are any anomolous RMSDS.

```python
import pandas as pd
from pathlib import Path
import IMP 
import IMP.atom
import IMP.rmf
import RMF
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
gsms_dir = Path("/wynton/group/sali/mhancock/mtorc2/exp_13/116/analysis/693502/sampcon_4/gsms")

rmf_file_clust = Path("/wynton/group/sali/mhancock/mtorc2/exp_13/116/analysis/693502/sampcon_4/1/cluster.0/cluster_center_model.rmf3")
fh_clust = RMF.open_rmf_file_read_only(str(rmf_file_clust))
m_clust = IMP.Model()
h_clust = IMP.rmf.create_hierarchies(fh_clust, m_clust)[0]
    
```

```python
for file in ["A", "B"]: 
    print(file)
    rmf_file_edit = Path(gsms_dir, "1000_edit/{}.rmf3".format(file))
    fh_out = RMF.create_rmf_file(str(rmf_file_edit))

    rmf_file = Path(gsms_dir, "1000/{}.rmf3".format(file))
    fh = RMF.open_rmf_file_read_only(str(rmf_file))
    
    txt_file_edit = Path(gsms_dir, "1000_edit/{}.txt".format(file))
    fh_txt_out = open(txt_file_edit, "w")
    
    txt_file = Path(gsms_dir, "1000/{}.txt".format(file))
    fh_txt = open(txt_file, "r")
    
    m = IMP.Model()
    h = IMP.rmf.create_hierarchies(fh, m)[0]

    IMP.rmf.add_hierarchy(fh_out, h)

    # for mol in ["MTOR", "RICTOR", "MLST8", "MSIN1", "CRIM", "RBD", "MSIN1PH", "AKT1PH", "KINASE"]:
    for i in range(1000):
        IMP.rmf.load_frame(fh, RMF.FrameID(i))
        score = float(fh_txt.readline())
        
        sel_1 = IMP.atom.Selection(h_clust, resolution=1, molecule="AKT1", copy_index=0)
        sel_2 = IMP.atom.Selection(h, resolution=1, molecule="AKT1", copy_index=0)

        pids_1 = [pid for pid in sel_1.get_selected_particle_indexes()]
        pids_2 = [pid for pid in sel_2.get_selected_particle_indexes()]

        xyzs_1 = [IMP.core.XYZ(m_clust, pid) for pid in pids_1]
        xyzs_2 = [IMP.core.XYZ(m, pid) for pid in pids_2]

    #         t = IMP.core.get_transformation_aligning_first_to_second(
    #             xyzs_2, xyzs_1)

        rmsd = IMP.atom.get_rmsd(xyzs_2, xyzs_1)
        if rmsd > 200: 
            print(i, score, rmsd)
        else: 
            print(i, score)
            IMP.rmf.save_frame(fh_out)
            fh_txt_out.write(str(score) + "\n")
            
    fh_txt_out.close()

    del fh
    del fh_out
    
```
